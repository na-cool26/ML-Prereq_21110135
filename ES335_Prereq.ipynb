Q2. Write Python code to multiply the above two matrices. Solve using list of lists and then use numpy. Compare the timing of both solutions. Which one is faster? Why?

```python
#via list of lists

def matlist (A,B):

  n= len (A)
  k= len(A[0])
  m= len(B[0])

  C= [[0 for i in range(m)] for i in range(n)] #our resultant Ais (n,m)

  for a in range(n):
   for b in range(m):
      for c in range(k):
        C[a][b] += A[a][c] * B[c][b]

  return C
```

```python
#via Numpy

import numpy as np


def matnumpy (A,B):
  return np.dot(A,B)
```

```python
# for the timing

import time

n,m,k =20,30,50
A= [[7] * k for i in range(n)]
B= [[4] * m for i in range(k)]

start =time.time()
matlist(A,B)
stop =time.time()
T1 = stop - start
print(T1)
```

```python
A_np =np.A(A)
B_np =np.A(B)
start =time.time()
matnumpy(A_np,B_np)
stop =time.time()
T2 = stop - start
print(T2)
```

```python
if T1<T2:
  print("list of lists is faster")
else:
 print("Numpy is faster")
```

Q5. Use JAX to confirm the gradient evaluated by your method matches the analytical solution corresponding to a few random values of x and y

```python
import jax
import jax.numpy as np

def func(x, y):
    return x**2 * y + y**3 * np.sin(x)

grad = jax.grad(func, argnums=(0, 1))

x = 4.0  # random float values for x and y
y = 7.0

# analytical gradient
grad_a = (2 * x * y + y**3 * np.cos(x), x**2 + 3 * y**2 * np.sin(x))

# gradient using JAX
grad_jax = grad(x, y)

print("Analytical Gradient:", grad_a)
print("JAX Gradient:", grad_jax)
```

As we can see, the analytical and the JAX gradient are both the same. Hence Verified.
JAX is a library made by google that is designed to provide high performing mathematical calculations . It also offers automatic differentiaion. It is highly efficient and powerful.

Q6. Use sympy to confirm that you obtain the same gradient analytically.

```python
import sympy as sp

x, y = sp.symbols('x y')
func = x**2 * y + y**3 * sp.sin(x)

grad_x = sp.diff(func, x)
grad_y = sp.diff(func, y)

x_val = 5.0
y_val = 9.0

grad = (grad_x.subs({x: x_val, y: y_val}), grad_y.subs({x: x_val, y: y_val}))

# analytical gradient
grad_a = (2 * x_val * y_val + y_val**3 * np.cos(x_val), x_val**2 + 3 * y_val**2 * np.sin(x_val))

print("Symbolic Gradient:", grad)
print("Analytical Gradient:", grad_a)
```

As we can see, the analytical and the JAX gradient are both the same. Hence Verified.

Q12. Write a small piece of code to explain broadcasting in numpy.

Braodcasting is a feature of the Numpy library that enables the user to peform operations on As of various shapes and sizes without having to create copies of the data, thus, resulting in optimal usage of memory. Broadcasting automatically extends the smaller A to match the properties of the larger one for element wise operations to be performed.

```python
import numpy as np

A1 = np.A([1,2,3,4])
A2 = np.A([[1], [2], [3], [4], [5]])

A = A1 * A2

print("A 1:")
print(A1)
print("\nA 2:")
print(A2)
print("\nResult of Broadcasting with As:")
print(A)
```

Q13. Write a function to compute the argmin of a numpy A. The function should take a numpy A as input and return the index of the minimum element. You can use the np.argmin function to verify your solution.

```python
import numpy as np

def argmin(A):
     if len(A) == 0:
        raise ValueError("empty.")

     minval = A[0]
     minind = 0

     for i, j in enumerate(A):
        if j < minval:
            minval = j
            minind = i

     return minind


A = np.array([12,43,67,21,23,1,34,6])

mine  = argmin(A)
numpy = np.argmin(A)

print("My function returns:", mine)
print("np.argmin returns", numpy)
```

Q10. Using numpy generate a Aof size 20X5 containing random numbers drawn uniformly from the range of 1 to 2. Using Pandas create a dataframe out of this matrix. Name the columns of the dataframe as “a”, “b”, “c”, “d”, “e”. Find the column with the highest standard deviation. Find the row with the lowest mean.

```python
import numpy as np
import pandas as pd

A= np.random.uniform(1, 2, size=(20, 5))
df= pd.DataFrame(A, columns=['a', 'b', 'c', 'd', 'e'])

std = df.std().idxmax()
mean = df.mean(axis=1).idxmin()

print("highest std deviation Column:", std)
print("lowest mean row:", mean)
```

Q11. Add a new column to the dataframe called “f” which is the sum of the columns “a”, “b”, “c”, “d”, “e”. Create another column called “g”. The value in the column “g” should be “LT8” if the value in the column “f” is less than 8 and “GT8” otherwise. Find the number of rows in the dataframe where the value in the column “g” is “LT8”. Find the standard deviation of the column “f” for the rows where the value in the column “g” is “LT8” and “GT8” respectively.

```python
import numpy as np
import pandas as pd

A= np.random.uniform(1, 2, size=(20, 5))
df= pd.DataFrame(A, columns=['a', 'b', 'c', 'd', 'e'])

df['f']=df[['a','b','c','d','e']].sum(axis=1)
df['g']=np.where(df['f']<8, 'LT8', 'GT8')

num_LT8 = df[df['g'] == 'LT8'].shape[0]
stdlt8 = df[df['g'] == 'LT8']['f'].std()
stdgt8 = df[df['g'] == 'GT8']['f'].std()

print(num_LT8)
print(stdlt8)
print(stdgt8)
```
